```python
import datetime, calendar
%mkdir output

data = []
data_pro_svatky = []
for kolikaty_mesic in range(1,13):

    year = 2020
    month = kolikaty_mesic
    num_days = calendar.monthrange(year, month)[1]
    days = [datetime.date(year, month, day) for day in range(1, num_days + 1)]

    datum = ''
    datum1 = []

    for day in range(1, num_days + 1):
        datum = f'{day}. {month}. {year}'
        datum1.append(datum)


    dny_v_mesici = []

    for i in datum1:
        den, mesic, rok = i.split('.')
        den_v_mesici = calendar.weekday(int(rok), int(mesic), int(den))
        dny_v_mesici.append(den_v_mesici)


    nazvy_dnu_v_mesici = []
    for i in dny_v_mesici:
        if i == 0:
            nazev_dne = 'po'
            nazvy_dnu_v_mesici.append(nazev_dne)
        elif i == 1:
            nazev_dne = 'út'
            nazvy_dnu_v_mesici.append(nazev_dne)
        elif i == 2:
            nazev_dne = 'st'
            nazvy_dnu_v_mesici.append(nazev_dne)
        elif i == 3:
            nazev_dne = 'čt'
            nazvy_dnu_v_mesici.append(nazev_dne)
        elif i == 4:
            nazev_dne = 'pá'
            nazvy_dnu_v_mesici.append(nazev_dne)
        elif i == 5:
            nazev_dne = 'so'
            nazvy_dnu_v_mesici.append(nazev_dne)
        elif i == 6:
            nazev_dne = 'ne'
            nazvy_dnu_v_mesici.append(nazev_dne)

    
    for i in range(0, calendar.monthrange(2020, month)[1]):
        #print(f'{nazvy_dnu_v_mesici[i]} {datum1[i]}')
        data.append(f'{nazvy_dnu_v_mesici[i]};{datum1[i]}') 
        
    data_pro_svatky.append(datum1)
    
```
